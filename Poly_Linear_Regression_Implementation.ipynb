```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# import  dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
# Splitting dats into Independent and dependent varibles
x=ds[['Level']]
y=ds[['Salary']]
```

```python
# plot the graphs
plt.scatter(x,y)
plt.show()
```

```python
# Model Fitting(Poly Linear Regression)
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
x_poly=poly_reg.fit_transform(x)
x_poly
reg=LinearRegression()
reg.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.transform(x)),color='purple')
plt.show()
```

```python
# Predicting my input

myinput=pd.DataFrame(data=[[11]], columns=["Level"])
result=reg.predict(poly_reg.transform(myinput))
result
```
