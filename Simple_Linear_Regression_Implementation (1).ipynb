```python
# import libraries
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# import dataset
ds=pd.read_csv("Salary_Data.csv")
ds
```

```python
# Divide dataset as Independent(Input) ans Dependent(output) varibles
x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
ds.columns
```

```python
x
```

```python
y
```

```python
# plot graph
plt.scatter(x,y)
plt.show()
```

```python
# Splitting dataset as Training and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Define model 
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train) # fix LinearRegression on training model
```

```python
# Predict 
result=reg.predict([[8]])
result
```

```python
result=reg.predict(x_test)
result
```

```python
y_test
```

```python
# Visulize
plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='gold')
plt.show()
```

```python
# for user input 
years= int(input("Enter Years of Experience:"))
my_input=pd.DataFrame(columns=['YearsExperience'],data=[[years]])
result=reg.predict(my_input)
print("Your Salary must be close to ",round(result[0,0]))
```
